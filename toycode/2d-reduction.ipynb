```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
txt = open('dimensionality-reduction.txt','r')
txt = txt.readlines()
```

```python
data = np.zeros((2000,3),dtype=np.float32)
for idx, line in enumerate(txt):
    data[idx] = line.strip().split('\t')
```

```python
data
```

```python
plt.plot(data)
```

```python
# Code from https://matplotlib.org/3.1.1/gallery/mplot3d/scatter3d.html
from mpl_toolkits.mplot3d import Axes3D 

fig = plt.figure()
c_lst = [plt.cm.rainbow(a) for a in np.linspace(0.0, 1.0, len(data))]

ax = fig.add_subplot(111, projection='3d')
for idx, point in enumerate(data):
    ax.scatter(point[0], point[1], point[2], marker='o', color=c_lst[idx])

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
# Code from https://woolulu.tistory.com/43
from sklearn.decomposition import PCA
# PCA 모델을 생성
pca = PCA(n_components=2) # 주성분 갯수
pca.fit(data) # PCA 적용
# 처음 두 개의 주성분으로 숫자 데이터를 변환
data_pca = pca.transform(data) # PCA를 데이터에 적용
plt.scatter(data_pca[::,0],data_pca[::,1])
```

```python
### TSNE모델에는 transform 메소드가 없고 fit_transform만 있음
# library import
from sklearn.manifold import TSNE
# t-SNE 모델 생성 및 학습
tsne = TSNE(random_state=0)
tsne_data = tsne.fit_transform(data)
plt.scatter(tsne_data[::,0],tsne_data[::,1])
```
